```python
from pyspark.sql import SparkSession
# Inicialize a sessão Spark
spark = SparkSession.builder \
    .appName("Consulta API em Spark") \
    .getOrCreate()

import pyspark.sql.functions as F
#from pyspark.sql.functions import explode, split, trim, col, regexp_replace #selecionar ou manipular colunas em operações de transformação e modificar dados de uma coluna
import requests #biblioteca para trazer dados de uma URL
import json #biblioteca para ler json
from sqlalchemy import create_engine #criar sessão no postgres
```

```python
# URL da API
url = "https://dadosabertos.compras.gov.br/modulo-material/1_consultarGrupoMaterial"

# Fazendo a solicitação HTTP
response = requests.get(url)

# Verifica se a solicitação foi bem sucedida
if response.status_code == 200:
    # Convertendo a resposta para JSON
    data = response.json()
    
    # Acessando a chave 'resultado' para obter os dados
    data_resultado = data['resultado']
    
    # Criando DataFrame Spark a partir dos dados
    GrupoMaterial = spark.createDataFrame(data_resultado)
    
    # Exibindo o DataFrame
    GrupoMaterial.show()
else:
    print("Falha ao obter os dados da API:", response.status_code)
```
